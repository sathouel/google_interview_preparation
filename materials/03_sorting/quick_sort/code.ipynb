# Using recursion

```python
def partition(tab, l, h):
    if l == h:
        return l
    
    pi = tab[l]
    lower = []
    higher = []
    skip = True
    for val in tab[l: h+1]:
        if val == pi and skip:
            skip = False
            continue
        if val > pi:
            higher.append(val)
        else:
            lower.append(val)
            
    new_pi_index = len(lower) if lower else 0
    tab[l: h+1] = lower + [pi] + higher
    return l + new_pi_index
```

```python
def quick_sort(tab, l, h):
    if h > l:
#         print("l == %d, h == %d" % (l, h))
        p_index = partition(tab, l, h)
        quick_sort(tab, l, p_index - 1)
        quick_sort(tab, p_index + 1, h)
```

```python
myList = [38, 27, 43, 3, 9, 82, 10]
  
quick_sort(myList, 0, len(myList) - 1)
myList
    
```

# Testing

```python
from random import randint

class quickSortTest:
    
    def __init__(self, list_number, list_max_size, list_max_val):
        self.list_number = list_number
        self.list_max_size = list_max_size
        self.list_max_val = list_max_val
        self.list_of_lists = []
        self.generate_lists()
        
    def generate_lists(self):
        lists_sizes = [randint(1, self.list_max_size) for _ in range(self.list_number)]
        for size in lists_sizes:
            l = [randint(0, self.list_max_val) for _ in range(size)]
            self.list_of_lists.append(l)
    
    def test_1(self):
        for i, l in enumerate(self.list_of_lists):
            sorted_1 = list(l)
            quick_sort(sorted_1, 0, len(l) - 1)
            l.sort()
            if sorted_1 != l:
                print("%d - error !" % i)
            else:
                print("%d - OK" % i)
    
```

```python
test = quickSortTest(200, 200, 200)
test.test_1()
```
